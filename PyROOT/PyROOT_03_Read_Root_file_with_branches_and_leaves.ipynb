# Read a root file with brancheas and leaves

We use and example of a Delphes root file which have branches and leaves

```python
import ROOT
from ROOT import TFile, TTree, TChain, TBranch, TLeaf
```

```python
# address of a root file
filename="/home/javad/Desktop/axaj_analysis/primary_root_file.root"

# read
rf = TFile(filename)
tr = rf.Get("Delphes")
```

## List Of Brancheas and Leaves

### List of Branches

```python
list_of_branches = []
for b in tr.GetListOfBranches():
    list_of_branches.append(b.GetName())
print(list_of_branches)
```

### List of Leaves

```python
list_of_leaves = []
for b in tr.GetListOfLeaves():
        list_of_leaves.append(b.GetName())
print(list_of_leaves)
```

## Read branches 

```python
branch = "Jet"
leaf   = "Jet.PT"
br = tr.GetBranch("Jet_size")
lf = tr.GetLeaf("Jet.PT")
```

```python
n_entries = tr.GetEntries()
n_entries = 1000
for event in xrange(n_entries):
    tr.GetEntry(event)
    for jet in xrange(tr.Jet_size):
        PT = tr.GetLeaf("Jet.PT").GetValue(jet)
        print(PT)
```

## Other method

```python
import ROOT
filename="/home/javad/Desktop/axaj_analysis/primary_root_file.root"

#================ Extracting Data from Delphes root =====================#
directory= filename
File = ROOT.TChain("Delphes;1")
File.Add(directory)
Number = File.GetEntries()

for i in range(Number):
    Entry = File.GetEntry(i)
    EntryFromBranch = File.Jet.GetEntries()
    for j in range(min(EntryFromBranch,2)):
        PT = File.GetLeaf("Jet.PT").GetValue(j)
        print(PT)
    print("==========")
```
