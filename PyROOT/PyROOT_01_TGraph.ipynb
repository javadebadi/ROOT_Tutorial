```python
import ROOT
from array import array
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
x = array('d',x)
y = array('d',y)
```

```python
g = ROOT.TGraph(len(x),x,y)
```

```python
c = ROOT.TCanvas()
c.cd()
g.Draw()
g.SetLineColor(2)
g.SetLineWidth(3)
g.SetMarkerStyle(10)
g.SetMarkerColor(4)
c.SetTickx()
c.SetTicky()
c.Draw()
```
