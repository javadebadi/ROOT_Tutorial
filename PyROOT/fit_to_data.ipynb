# fit to data

```python
import ROOT
from ROOT import TF1, TGraph, TCanvas
from array import array
```

```python
def fit_now(ym,yp):
    x = [-0.1,0,1.00]
    y = [ym,0.14,yp]
    x = array('d',x)
    y = array('d',y)
    gr = TGraph(3,x,y)
    f = TF1("f", "[1] * x * x + [0] * x + 0.14"); 
    gr.Fit(f);
    print(f.Eval(-1))
```

```python
def fit_now_CP(ym,yp):
    x = [-0.1,0,1.00]
    y = [ym,0.14,yp]
    x = array('d',x)
    y = array('d',y)
    gr = TGraph(3,x,y)
    f = TF1("f", "[0] * x * x + 0.14"); 
    gr.Fit(f);
    print(f.Eval(-1))
```
